```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_facecolor('black')

t = np.linspace(0, 2*np.pi, 101)
plt.plot(1.1*np.cos(t), 1.1*np.sin(t), 'k')
plt.plot(1.2*np.cos(t), 1.2*np.sin(t), 'k')


plt.plot((1.3)*np.cos(t), (1.3)*np.sin(t), 'g')
plt.plot((1.4)*np.cos(t), (1.4)*np.sin(t), 'c')
plt.plot((1.5)*np.cos(t), (1.5)*np.sin(t), 'm')
plt.plot((1.6)*np.cos(t), (1.6)*np.sin(t), 'pink')

plt.plot(1.7*np.cos(t), 1.7*np.sin(t), 'k')#外圍紅圈
plt.plot(1.8*np.cos(t), 1.8*np.sin(t), 'k')

TEXT = 'NTHUNCTU'#白圈文字(唯一支持清交小徑)
for i in range(len(TEXT)):
    s = np.pi*2*i/len(TEXT)
    x0,y0 = 1.45*np.cos(s), 1.45*np.sin(s)
    ax.text(x0, y0, TEXT[i], 
            color='white', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontsize=16)
    plt.plot(.15*np.cos(t)+x0, .15*np.sin(t)+y0, 'w')
    
for i in range(8):#紅圈
    t0 = 2*np.pi/8*i
    x0,y0 = 1.45*np.cos(t0), 1.45*np.sin(t0)
    plt.plot(.3*np.cos(t)+x0, .3*np.sin(t)+y0, 'r')
plt.plot(np.cos(t), np.sin(t), 'r:')
plt.plot(np.cos(t+np.pi/4), np.sin(t+np.pi/4), 'r:')

t = np.linspace(0, 2*np.pi, 211)
plt.plot(1.15*np.cos(t), 1.15*np.sin(t), 'yp')#黃圈

t = np.linspace(0, 2*np.pi, 211)
plt.plot(1.75*np.cos(t), 1.75*np.sin(t), 'yp')#黃圈

t = np.linspace(0, 4*np.pi, 6)
plt.plot(1.0*np.cos(t+np.pi/10), 1.0*np.sin(t+np.pi/10), 'y')#星星


t = np.linspace(0, 2*np.pi, 10)
plt.plot(np.cos(t+np.pi/4), np.sin(t+np.pi/4), 'r:')#十邊形

t = np.linspace(0, 2*np.pi, 5)
plt.plot(np.cos(t), np.sin(t), 'r:')#菱形


    
plt.savefig('circle.png')

```
